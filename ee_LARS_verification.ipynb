```python
import scipy
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy
import pandas
import sklearn
from sklearn import preprocessing
from sklearn import linear_model
import tensorflow
```

```python
inputCollection = pandas.read_csv('Test4.csv')
X = inputCollection.drop(labels='LAI', axis=1)
y = inputCollection.LAI
```

```python
LAImodel = sklearn.linear_model.Lars(n_nonzero_coefs=13)
LAImodel = LAImodel.fit(X, y)
```

```python
print('LAI Coefficient Path')
pandas.DataFrame(LAImodel.coef_path_).loc[:, 0:13]
```
